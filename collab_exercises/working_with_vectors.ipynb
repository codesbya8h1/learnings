Given 3D Vectors [40, 120, 60] and [60, 50, 90]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Initialize the vectors

color1 = [40, 120, 60]
color2 = [60, 50, 90]

ax = plt.figure().add_subplot(projection='3d')

#Lets draw these vectors

ax.scatter(xs=color1[0], ys=color1[1], zs=color1[2], zdir='z', label='color1')
ax.scatter(xs=color2[0], ys=color2[1], zs=color2[2], zdir='z', label='color1')

#origin [0,0,0]

ax.scatter(0,0,0, zdir='z', label='origin', c='black')

ax.quiver(0,0,0, color1[0], color1[1], color1[2], color="black", arrow_length_ratio=0.1)
ax.quiver(0,0,0, color2[0], color2[0], color2[2], color="black", arrow_length_ratio=0.1)

# Make legend, set axes limits and labels

ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#Customize the view angle so its easier to see that the scatter points line on the plane y=0
ax.view_init(elev=20, azim=-35, roll=0)
plt.show()

```

```python
# Calculate the euclidian, manhattan, cosine, and  dot product distances between these vectors

# Manhattan Distance

L1 = [color1[i] - color2[i] for i in range(len(color1))]
L1 = np.abs(L1).sum()
print(L1)
```

```python
# Euclidian distance

L2 = [(color1[i]-color2[i])**2 for i in range(len(color1))]
L2 = np.sqrt(np.array(L2).sum())
print(L2)
```

```python
# Cosine Distance

a = np.array(color1)
b = np.array(color2)
cosine_dist = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print(cosine_dist)
```

```python
# Dot product

dot_product = np.array([color1[i]*color2[i] for i in range(len(color1))]).sum()
print(dot_product)
```

```python
#Initialize the vectors

color1 = [40, 120, 60]
color2 = [60, 50, 90]
color3 = [color1[0]+10, color1[1]+15, color1[2]+5]

ax = plt.figure().add_subplot(projection='3d')

#Lets draw these vectors

ax.scatter(xs=color1[0], ys=color1[1], zs=color1[2], zdir='z', label='color1')
ax.scatter(xs=color2[0], ys=color2[1], zs=color2[2], zdir='z', label='color2')

#origin [0,0,0]

ax.scatter(0,0,0, zdir='z', label='origin', c='black')

ax.quiver(0,0,0, color1[0], color1[1], color1[2], color="black", arrow_length_ratio=0.1)
ax.quiver(0,0,0, color2[0], color2[1], color2[2], color="black", arrow_length_ratio=0.1)

#Add another vector to the diagram
ax.scatter(xs=color3[0], ys=color3[1], zs=color3[2], zdir='z', label='NewColor')
ax.quiver(0,0,0, color3[0], color3[1], color3[2], color="black", arrow_length_ratio=0.1)


# Make legend, set axes limits and labels

ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#Customize the view angle so its easier to see that the scatter points line on the plane y=0
ax.view_init(elev=20, azim=-35, roll=0)
plt.show()

```
